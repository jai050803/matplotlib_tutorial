# Matplotlib Library

#### What is a bubble chart?

A bubble chart is a type of data visualization that displays three dimensions of data using bubbles of varying sizes. It is similar to a scatter plot, as both types of charts use x and y axes to represent two variables. However, in a bubble chart, the third variable is represented by the size of the bubbles.

Here are the key components of a bubble chart:

X-axis: This represents one variable of the data being analyzed.

Y-axis: This represents another variable of the data being analyzed.

Bubble size: This represents the third variable. Larger bubbles indicate higher values, while smaller bubbles indicate lower values for this variable.

Color (optional): Some bubble charts also use color to represent a fourth variable. This adds an additional layer of information to the chart.

##### Importing the necessary libraries

```python
import matplotlib.pyplot as plt 
import numpy as np
```

Creating a bubble chart. We first create two numpy arrays for x_values and y_values, then using the matplotlib function of plotting the bubble chart to plot the graph

```python
x_values = np.array([1,2,3,4,5,6])
y_values = np.array([12,17,24,10,6,30])
bubble_size = np.array([105,106,107,108,109,110])
color = ['red', 'green', 'blue','yellow', 'black', 'hotpink']

plt.scatter(x_values, y_values, s=bubble_size, c=color, alpha=0.7)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("bubble chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for the example
np.random.seed(0)
x = np.random.rand(20) * 10  # X-axis data
y = np.random.rand(20) * 100  # Y-axis data
bubble_sizes = np.random.rand(20) * 1000  # Bubble size data
colors = np.random.rand(20)  # Color data

# Create the bubble chart
plt.scatter(x, y, s=bubble_sizes, c=colors, alpha=0.5)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bubble Chart Example')

# Add colorbar legend
plt.colorbar(label='Color')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for the example
np.random.seed(0)
x = np.random.rand(20) * 10  # X-axis data
y = np.random.rand(20) * 100  # Y-axis data
bubble_sizes = np.random.rand(20) * 1000  # Bubble size data

# Generate random RGB colors
colors = np.random.rand(20, 3)  # 20 rows, 3 columns for RGB

# Create the bubble chart with random colors
plt.scatter(x, y, s=bubble_sizes, c=colors, alpha=0.5)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bubble Chart with Random Colors')

# Show the plot
plt.show()
```

##### Perform plotting using this array y=np.array([10,5,18,20]), by atleast 5 attributes.


```python
# Define the array
y = np.array([10, 5, 18, 20])

# Line plot
plt.subplot(231)
plt.plot(y, marker='o', linestyle='-', color='b')
plt.title('Line Plot')
plt.xlabel('Index')
plt.ylabel('Values')

# Scatter plot
plt.subplot(232)
plt.scatter(range(len(y)), y, color='r', s=100, alpha=0.8)
plt.title('Scatter Plot')
plt.xlabel('Index')
plt.ylabel('Values')

# Bar plot
plt.subplot(233)
plt.bar(range(len(y)), y, color='g', alpha=0.6)
plt.title('Bar Plot')
plt.xlabel('Index')
plt.ylabel('Values')

# Histogram
plt.subplot(234)
plt.hist(y, bins=3, color='y', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')

# Pie chart
plt.subplot(235)
plt.pie(y, labels=['A', 'B', 'C', 'D'], autopct='%1.1f%%', colors=['cyan', 'magenta', 'yellow', 'orange'])
plt.title('Pie Chart')

plt.tight_layout()
plt.show()
```
