# Matplotlib Library

#### What is a bubble chart?

A bubble chart is a type of data visualization that displays three dimensions of data using bubbles of varying sizes. It is similar to a scatter plot, as both types of charts use x and y axes to represent two variables. However, in a bubble chart, the third variable is represented by the size of the bubbles.

Here are the key components of a bubble chart:

X-axis: This represents one variable of the data being analyzed.

Y-axis: This represents another variable of the data being analyzed.

Bubble size: This represents the third variable. Larger bubbles indicate higher values, while smaller bubbles indicate lower values for this variable.

Color (optional): Some bubble charts also use color to represent a fourth variable. This adds an additional layer of information to the chart.

##### Importing the necessary libraries

```python
import matplotlib.pyplot as plt 
import numpy as np
```

Creating a bubble chart. We first create two numpy arrays for x_values and y_values, then using the matplotlib function of plotting the bubble chart to plot the graph

```python
x_values = np.array([1,2,3,4,5,6])
y_values = np.array([12,17,24,10,6,30])
bubble_size = np.array([105,106,107,108,109,110])
color = ['red', 'green', 'blue','yellow', 'black', 'hotpink']

plt.scatter(x_values, y_values, s=bubble_size, c=color, alpha=0.7)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("bubble chart")
plt.show()
```
