# Matplotlib programs

importing numpy and matplotlib pyplot library

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,5,7,9,11])

plt.plot(x,y,marker = "d")

plt.show()
```

## bar graph

```python
x = np.array([1,2,3,4,5])
y = np.array([2,5,7,9,11])

plt.bar(x,y)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,5,7,9,11])

plt.scatter(x,y)
plt.show()
```

## Area plot

```python
x = [1,2,3,4,5]
y1 = [20,40,60,80,100]
y2 = [30,50,70,90,50]

plt.fill_between(x,y1,y2, alpha=0.5)
plt.show()
```
