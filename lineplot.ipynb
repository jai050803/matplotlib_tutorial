# Matplotlib programs

importing numpy and matplotlib pyplot library

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,5,7,9,11])

plt.plot(x,y,marker = "d")

plt.show()
```

## bar graph

```python
x = np.array([1,2,3,4,5])
y = np.array([2,5,7,9,11])

plt.bar(x,y)
plt.show()
```
