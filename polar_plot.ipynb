## Polar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
theta = np.linspace(0, 2 * np.pi, 100)
r = np.sin(theta)
```

```python
plt.subplot(111, projection='polar')
plt.plot(theta, r)
plt.show()
```
