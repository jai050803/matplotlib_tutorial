## Polar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
theta = np.linspace(0, 2 * np.pi, 100)
r = np.sin(theta)
```

```python
plt.subplot(111, projection='polar')
plt.plot(theta, r)
plt.show()
```

## Waffle Plot

```python
from pywaffle import Waffle
import pandas as pd
```

```python
data =  {'phone' : ['a','b','c','d','e'],
         'stock': [44,12,8,5,3]
        }

df = pd.DataFrame(data)

fig = plt.figure(FigureClass= Waffle, rows=5, values=df.stock,label=list(df.phone))
plt.show()
```

## Word Cloud

```python
from wordcloud import WordCloud
```

```python
text = "Hey i am jai and i am a btech data science student"

word_cloud = WordCloud(width=500,height=500,background_color="black").generate(text)
plt.figure(figsize=(10,5))
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis=('off')
plt.title('word Cloud')
plt.show()
```

## 3D Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0,1,100)
x= z * np.sin(25*z)
y= z * np.cos(25*z)
c= x+y
ax.scatter(x,y,z, c=c)
ax.set_title("3D plot")
plt.show()
```
