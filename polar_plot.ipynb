## Polar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
theta = np.linspace(0, 2 * np.pi, 100)
r = np.sin(theta)
```

```python
plt.subplot(111, projection='polar')
plt.plot(theta, r)
plt.show()
```

## Waffle Plot

```python
from pywaffle import Waffle
import pandas as pd
```

```python
data =  {'phone' : ['a','b','c','d','e'],
         'stock': [44,12,8,5,3]
        }

df = pd.DataFrame(data)

fig = plt.figure(FigureClass= Waffle, rows=5, values=df.stock,label=list(df.phone))
plt.show()
```

## Word Cloud

```python
from wordcloud import WordCloud
```

```python
text = "Hey i am jai and i am a btech data science student"

word_cloud = WordCloud(width=500,height=500,background_color="black").generate(text)
plt.figure(figsize=(10,5))
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis=('off')
plt.title('word Cloud')
plt.show()
```

## 3D Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0,1,100)
x= z * np.sin(25*z)
y= z * np.cos(25*z)
c= x+y
ax.scatter(x,y,z, c=c)
ax.set_title("3D plot")
plt.show()
```

## Contour Plot

```python
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X, Y = np.meshgrid(x,y)

Z = np.sin(X) + np.cos(Y)

contour_plot = plt.contour(X,Y,Z)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Contour plot example")

plt.colorbar(contour_plot, label = 'Z')
plt.show()
```

```python
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X, Y = np.meshgrid(x,y)

Z = X**2 + Y**2

contour_plot = plt.contour(X,Y,Z)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Contour plot example")

plt.colorbar(contour_plot, label = 'Z')
plt.show()
```

```python
angles = np.linspace(0, 2*np.pi, 6, endpoint=False)
# radii = [1,2,3,4,5,6]

#creating a polar plot
plt.polar(angles,marker='o',linestyle='-', label= "polar plot")
plt.legend()
plt.show()
```

```python

# Define 10 equally spaced angles from 0 to 2*pi
angles = np.linspace(0, 2*np.pi, 10, endpoint=False)

# Define random radii values for each angle (you can replace this with your own data)
radii = np.random.randint(1, 10, size=10)

# Create a polar plot with markers at given angles and radii
plt.figure(figsize=(8, 6))  # Adjust the figure size if needed
plt.polar(angles, radii, marker='o', linestyle='-',markersize=8,markeredgewidth=2,markeredgecolor='black',markerfacecolor="hotpink", label="Polar Plot")

# Add a legend to the plot
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.8)

plt.fill(angles, radii, color='lightblue', alpha=0.3)
plt.ylim(0, 10)



# Display the polar plot
plt.show()
```
