## Polar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
theta = np.linspace(0, 2 * np.pi, 100)
r = np.sin(theta)
```

```python
plt.subplot(111, projection='polar')
plt.plot(theta, r)
plt.show()
```

## Waffle Plot

```python
from pywaffle import Waffle
import pandas as pd
```

```python
data =  {'phone' : ['a','b','c','d','e'],
         'stock': [44,12,8,5,3]
        }

df = pd.DataFrame(data)

fig = plt.figure(FigureClass= Waffle, rows=5, values=df.stock,label=list(df.phone))
plt.show()
```

## 3D Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0,1,100)
x= z * np.sin(25*z)
y= z * np.cos(25*z)
c= x+y
ax.scatter(x,y,z, c=c)
ax.set_title("3D plot")
plt.show()
```

## Contour Plot

```python
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X, Y = np.meshgrid(x,y)

Z = np.sin(X) + np.cos(Y)

contour_plot = plt.contour(X,Y,Z)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Contour plot example")

plt.colorbar(contour_plot, label = 'Z')
plt.show()
```

```python

# Define 10 equally spaced angles from 0 to 2*pi
angles = np.linspace(0, 2*np.pi, 10, endpoint=False)

# Define random radii values for each angle (you can replace this with your own data)
radii = np.random.randint(1, 10, size=10)

# Create a polar plot with markers at given angles and radii
plt.figure(figsize=(8, 6))  # Adjust the figure size if needed
plt.polar(angles, radii, marker='o', linestyle='-',markersize=8,markeredgewidth=2,markeredgecolor='black',markerfacecolor="hotpink", label="Polar Plot")

# Add a legend to the plot
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.8)

plt.fill(angles, radii, color='lightblue', alpha=0.3)
plt.ylim(0, 10)



# Display the polar plot
plt.show()
```

```python
# Generate random data for illustration purposes
altitude = np.linspace(0, 5000, 100)
latitude = np.linspace(-90, 90, 100)
altitude, latitude = np.meshgrid(altitude, latitude)
temperature = np.sin(altitude/1000) + np.cos(latitude/10)
humidity = np.random.rand(100, 100) * 100
wind_speed = np.random.rand(100, 100) * 20
cloud_cover = np.random.rand(100, 100) * 100
pressure = np.random.randint(800, 1200, size=(100, 100))
solar_radiation = np.random.randint(100, 1000, size=(100, 100))
ozone_concentration = np.random.rand(100, 100) * 200
precipitation = np.random.rand(100, 100) * 50

# Create the contour plot
plt.figure(figsize=(10, 8))

# Contour plot for temperature
plt.subplot(3, 4, 1)
plt.contourf(altitude, latitude, temperature, cmap='coolwarm')
plt.colorbar()
plt.title('Temperature (C)')

# Contour plot for humidity
plt.subplot(3, 4, 2)
plt.contourf(altitude, latitude, humidity, cmap='Blues')
plt.colorbar()
plt.title('Humidity (%)')

# Contour plot for wind speed
plt.subplot(3, 4, 3)
plt.contourf(altitude, latitude, wind_speed, cmap='Oranges')
plt.colorbar()
plt.title('Wind Speed (m/s)')

# Contour plot for cloud cover
plt.subplot(3, 4, 4)
plt.contourf(altitude, latitude, cloud_cover, cmap='Greys')
plt.colorbar()
plt.title('Cloud Cover (%)')

# Contour plot for pressure
plt.subplot(3, 4, 5)
plt.contourf(altitude, latitude, pressure, cmap='RdYlBu')
plt.colorbar()
plt.title('Pressure (hPa)')

# Contour plot for solar radiation
plt.subplot(3, 4, 6)
plt.contourf(altitude, latitude, solar_radiation, cmap='YlOrBr')
plt.colorbar()
plt.title('Solar Radiation (W/m^2)')

# Contour plot for ozone concentration
plt.subplot(3, 4, 7)
plt.contourf(altitude, latitude, ozone_concentration, cmap='Purples')
plt.colorbar()
plt.title('Ozone Concentration (ppb)')

# Contour plot for precipitation
plt.subplot(3, 4, 8)
plt.contourf(altitude, latitude, precipitation, cmap='BuPu')
plt.colorbar()
plt.title('Precipitation (mm)')

plt.tight_layout()
plt.show()
```

```python

# Generate random data for illustration purposes
altitude = np.linspace(0, 5000, 100)
latitude = np.linspace(-90, 90, 100)
altitude, latitude = np.meshgrid(altitude, latitude)
temperature = np.sin(altitude / 1000) + np.cos(latitude / 10)

# Create the contour plot using contourf
plt.figure(figsize=(12, 6))

# Filled contour plot with contourf
plt.subplot(1, 2, 1)
contourf = plt.contourf(altitude, latitude, temperature, cmap='coolwarm')
plt.colorbar(contourf)
plt.title('Filled Contour Plot (contourf)')

# Contour plot without filling using contour
plt.subplot(1, 2, 2)
contour = plt.contour(altitude, latitude, temperature, cmap='coolwarm')
plt.colorbar(contour)
plt.title('Contour Plot (contour)')

plt.tight_layout()
plt.show()
```

```python

# Sample text data
text = "Python is a popular programming language for data analysis and machine learning. Data visualization is an important aspect of data analysis. Matplotlib is a powerful library for creating visualizations in Python."

# Generate the word cloud with custom parameters
wordcloud = WordCloud(width=800, height=400, max_words=100, background_color='lightgray', colormap='viridis', 
                      contour_color='black', contour_width=1, font_path='arial', min_font_size=10).generate(text)

# Plot the word cloud with customizations
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
# plt.axis('off')
plt.show()
```
